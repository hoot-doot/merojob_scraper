```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def get_jobs(category, pages):
    '''Gets scrapes the job title, company name, location, views and skills'''
    job_list = []
    for no in range(1,pages):
        # goes through the given number of pages
        url = "https://merojob.com/category/" + category + "/?page="+ str(no)
        r = requests.get(url)
        soup = bs(r.content, 'html.parser')
        jobs = soup.find('div',attrs={'id':'search_job'})
        for x in range(0,5):

            # extracting data

            job_title = jobs.find_all(itemprop="title")[x].get_text(" ", strip=True)
            orgn = jobs.select('h3.h6')[x].get_text(" ", strip=True)
            location = jobs.find_all(itemprop="address")[x].get_text(" ", strip=True)
            views = jobs.find_all('span',attrs={'class':'text-primary mr-2'})[x].get_text(" ", strip=True).replace("Views: ","")
            try:
                desc = jobs.find_all(itemprop="skills")[x].get_text(" ", strip=True).split(" ")
            except:
                desc = ("No skills listed")

            # adding data to list

            job_list.append({"Job Title" : job_title,
            "Company" : orgn,
            "Location" : location,
            "Views" : int(views),
            "Skills" : desc})
            
    return pd.DataFrame(job_list)#changes the list into a pandas dataframe
```

```python
it = get_jobs("it-telecommunication", 10)#catagory and page number
(it)
```
